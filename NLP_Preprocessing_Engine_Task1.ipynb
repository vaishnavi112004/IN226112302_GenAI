# Task 1: Conceptual Understanding

## 1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing (NLP), "Love" and "love" are treated as two different tokens because text is case-sensitive by default. This means they are considered separate words in the vocabulary.

To avoid this issue, preprocessing usually includes converting all text to lowercase. After this step, both "Love" and "love" become the same word, which helps reduce duplication and improves model performance.

## 2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset becomes noisy and less meaningful. Common words like "is", "the", and "and" appear very frequently but do not add much value to understanding the text.

This can:

* Increase processing time
* Reduce model accuracy
* Make it harder to identify important words

---

## 3. Provide two real-world scenarios where removing stopwords can be harmful

1. Sentiment Analysis
   In sentences like "This is not good", removing the word "not" changes the meaning completely and gives the wrong sentiment.

2. Question Answering or Chatbots
   In a question like "Where is the nearest hospital?", removing words like "where" or "is" can affect the intent and make the question unclear.

## 4. Explain the difference between stemming and lemmatization

Stemming is a process that removes suffixes from words to get their root form. For example, "running" may become "run" or "runn". It is faster but less accurate.

Lemmatization converts a word to its base or dictionary form using linguistic rules. For example, "running" becomes "run". It is more accurate but slightly slower than stemming.


```python
# Task 2: Advanced Preprocessing Function

import re

def preprocess_text(text):

    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ("no", "not")]

    # Recreate cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence


# Example test
example = "WOW!!! This is soooo coool!!! Visit http://example.com now"
tokens, cleaned = preprocess_text(example)

print("Tokens:", tokens)
print("Cleaned sentence:", cleaned)
```

```python
# Task 3: Stress Testing
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```

```python
# Task 4: Token Analytics

def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total if total > 0 else 0

    return total, unique, avg_len


print("\n========== TOKEN ANALYSIS ==========")

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    total, unique, avg_len = token_analysis(tokens)

    print("\nOriginal sentence:", sentence)
    print("Cleaned sentence:", clean)
    print("Total tokens:", total)
    print("Unique tokens:", unique)
    print("Avg token length:", round(avg_len, 2))
```

```python
# Task 5: Frequency Analysis

from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
```

```python
# Task 6: Build Full Pipeline Function

def full_pipeline(text_list):
    """
    Processes a list of text inputs using preprocess_text()
    Returns:
    {
        "tokens": [...],
        "clean_sentences": [...]
    }
    """

    # Error handling: input must be a list
    if not isinstance(text_list, list):
        raise ValueError("Input must be a list of strings")

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, cleaned_sentence = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Example test run

pipeline_output = full_pipeline(test_sentences)

print("All Tokens:\n", pipeline_output["tokens"])
print("\nClean Sentences:\n", pipeline_output["clean_sentences"])
```

```python
# Task 7: Error Handling Tests

error_test_cases = [
    "",              # Empty string
    "😍🔥😂",         # Only emojis
    "1234567890"     # Only numbers
]

print("ERROR HANDLING TEST RESULTS\n")

for i, text in enumerate(error_test_cases, start=1):

    tokens, cleaned_sentence = preprocess_text(text)

    print(f"Test Case {i}")
    print("Original Text   :", repr(text))
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```
